```python
# importing required libraries
import pandas as pd
```

```python
# reading train data
train_features = pd.read_csv("/content/drive/MyDrive/cgas_project/train_features.csv")
```

```python
train_features
```

```python
train_labels = pd.read_csv("/content/drive/MyDrive/cgas_project/train_labels.csv")
```

```python
train_labels
```

```python
import tqdm
```

```python
# Making dictionary of different Recipe IDs
data_dict = {}
for index, row in train_features.iterrows():
  if row['Recipe_ID'] in data_dict:
    data_dict[row['Recipe_ID']][0].append(row['Ingredient_ID'])
    data_dict[row['Recipe_ID']][1].append(row['state'])
    data_dict[row['Recipe_ID']][2].append(row['quantity'])
    data_dict[row['Recipe_ID']][3].append(row['unit'])
  else:
    data_dict[row['Recipe_ID']] = [[],[],[],[]]
    data_dict[row['Recipe_ID']][0].append(row['Ingredient_ID'])
    data_dict[row['Recipe_ID']][1].append(row['state'])
    data_dict[row['Recipe_ID']][2].append(row['quantity'])
    data_dict[row['Recipe_ID']][3].append(row['unit'])

  print(index)
```

```python
data_dict[2610]
```

```python
df_result = pd.DataFrame()
```

```python
df_result['Recipe_ID'] = ''
df_result['Ingredient'] = ''
df_result['state'] = ''
df_result['quantity'] = ''
df_result['Unit'] = ''
```

```python
i = 0
for ele in data_dict.keys():
  df_result.loc[i,'Recipe_ID'] = ele
  df_result.loc[i,'Ingredient'] = data_dict[ele][0]
  df_result.loc[i,'state'] = data_dict[ele][1]
  df_result.loc[i,'quantity'] = data_dict[ele][2]
  df_result.loc[i,'Unit'] = data_dict[ele][3]
  i+=1
```

```python
df_result
```

```python
df_result.to_csv("result.csv",index=False)
```
